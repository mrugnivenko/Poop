```python
import numpy as np
from numpy import linalg
import math
from matplotlib import pylab as plt
from mpl_toolkits.mplot3d import Axes3D
%matplotlib notebook
```

```python
def surface_plot (matrix, **kwargs):
    # acquire the cartesian coordinate matrices from the matrix
    # x is cols, y is rows
    (x, y) = np.meshgrid(np.arange(matrix.shape[1]), np.arange(matrix.shape[0]))
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')
    surf = ax.plot_surface(x, y, matrix, **kwargs)
    return (fig, ax, surf)
```

```python
def plot_it(txt_file):
    u = read_dataset(txt_file)

    (fig, ax, surf) = surface_plot(u, cmap=plt.cm.coolwarm)

    fig.colorbar(surf)

    ax.set_xlabel('p')
    ax.set_ylabel('slaves (cores)')
    ax.set_zlabel('u (values)')

    plt.show()
```

```python
n_to_read = 6
d_to_read = 9
```

```python
def read_dataset(file_name):
    global n_to_read
    global d_to_read
    out = np.zeros((n_to_read, d_to_read))
    f = open(file_name, 'r')
    for i in range(n_to_read):
        s = f.readline()
        tokens = s.split(' ')
        line_to_write = []
        for j in range(d_to_read):
            tok_tok = []
            tok_tok = str(tokens[j]).split('_')
            line_to_write.append(tok_tok[1])
            out[i][j] = line_to_write[j]
    return out
```

```python
plot_it("plots8.txt")
```
